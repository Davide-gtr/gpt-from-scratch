## Read dataset

```python
file = "./input.txt"
with open(file, 'r', encoding='utf-8') as f:
    dataset = f.read()
```

```python
print(f"Length of the dataset = {len(dataset)}")
```

```python
# Printing first 100 characters
print(dataset[:100])
```

## Check all unique characters in the text

set: create a set with all characters present in the text  
list: cast as list  
sorted: sort it

```python
chars = sorted(list(set(dataset)))
vocab_size = len(chars)
print("All characters:", ''.join(chars))
print("Number of unique characters =", vocab_size)
```
